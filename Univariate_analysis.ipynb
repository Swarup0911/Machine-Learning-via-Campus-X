```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/csv files/train.csv')
```

```python
df.head()
```

# **1. Categorical Data**

# **a. Countplot**

```python
sns.countplot(df['Embarked'] , palette='deep')
#df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Pclass', data=df, palette='pastel')
# The palette parameter in Seaborn controls the colors used in plots.
# Especially when you have categories (like in bar plots, pie charts, etc.).
# Here Pastel is a predefined category in seaborn .
```

```python
sns.countplot(x='Survived' , data = df , palette = 'dark')
```

# **b. Pie-Chart**

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# **2. Numerical Data**

# **a. Histogram**

```python
import matplotlib.pyplot as plt
sns.histplot(df['Age'])
```

# **b. Dist-Plot**

```python
sns.distplot(df['Age'])#  Figure-level distribution plot (combines hist, KDE, ECDF).
```

```python
sns.kdeplot(df['Age']) #  Kernel Density Estimate (smooth distribution curve).
```

```python

sns.ecdfplot(df['Age'])#  Empirical Cumulative Distribution Function.

```

```python
sns.rugplot(df['Age'])#  Adds small ticks to show individual data points.
```

# **c. Box-Plot**

```python
sns.boxplot(df['Age'])
```
