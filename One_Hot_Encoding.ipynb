# **2. K-1 OneHotEncoding**

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
#Problem of Multicollinearity is solved.
```

**Pandas is not used for OHE while doing machine learning projects bcoz pandas does not remember the order of the data while splitting it into categories. So, we use Scikit Learn .**

# **3. OneHotEncoding using Sklearn**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False ,dtype=np.int32)
#drop usually drops the first category of splitting so, when we check the shape we will get 2 less columns
```

```python
# Here we are only working on fuel and owner as, the brand has 32 different categories, which will increase the dimensionality of the df
X_train_new = ohe.fit_transform(X_train[['fuel','owner']]) #.toarray()
#if we write sparse_output = False then we donot need to use .toarray and it would directly return np array.
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']]) #.toarray()
```

# **4. OneHotEncoding with Top Categories**

```python
df['brand'].value_counts()
```

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
counts[counts <= threshold].index
# storing the indexes of the brands that donot cross the threshold into repl
```

```python
repl = counts[counts <= threshold].index
# storing the indexes of the brands that donot cross the threshold into repl
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
#merging the brands with less cars into a single column named uncommon
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).sample(5)
```
